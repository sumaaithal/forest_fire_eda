```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_copy = df.copy()
```

```python
df_copy.head()
```

```python
df_copy.drop(['X','Y'],axis=1,inplace=True)
```

```python
df_copy.head()
```

ffmc : fine fuel moisture code under shade of  forest canopy
dmc : duff moisture code decomposed organic matter under the forest
dc : drought code , represents drought
isi : initial spread index, integrates fuel moisture for fine dead fuels causing fire
rh : relative humidity

```python
sns.barplot(x='month',y='DC',data=df_copy)
```

### it is concluded from the above plot that month sept have highest drought in the entire year.

```python
df_num = df_copy.copy()
```

```python
df_num.drop(['month','day'],axis=1,inplace=True)
```

```python
sns.distplot(df_num,hist=True)
```

```python
sns.pairplot(df_num)
```

# from the above pair plot it is concluded that 

#### temperature and humidity rate is positively correlated and it some-what has the normal distrubution in histogramic representation 

#### DMC(duff moisture code) &DC(drought code) are linearly(+ve) correlated to each other, as the DC increases, DMC also increases

#### Rain (less/constant) wind more(keeps on increasing)

#### average/small area, wind is high

```python
sns.barplot(x="day",y="temp",data=df_copy)
```

### from the above graph is is seen that, (tue,wed,thurs) as a week day has the highest temperature which could lead to forest-fire

```python
plt.figure(figsize=(10,6))
plt.subplot(121)
sns.lineplot(x="day",y="rain",data=df_copy)
plt.subplot(122)
sns.lineplot(x="month",y="rain",data=df_copy)
```

### from the lineplot it is noted that, chances of high rain is possible for month of  (apr,aug,dec) and (tues,wed,thur) in weekdays, from my opinion, there won't be a very high chance of forest fires during this time

```python
plt.figure(figsize=(10,8))
plt.subplot(121)
sns.barplot(x="month",y='RH',data=df_copy)

plt.subplot(122)
sns.barplot(x="day",y='RH',data=df_copy)
```

### as the conclusion, RH is high for weekdays (mon,fri,sun)  and  month(jan,feb,may)

```python
plt.figure(figsize=(10,8))
plt.subplot(121)
sns.barplot(x="month",y='ISI',data=df_copy)

plt.subplot(122)
sns.barplot(x="day",y='ISI',data=df_copy)
```

### ISI is high for weekdays(tues,wed,thu) and months(feb,jun,aug) and these attributes might cause the forestfire

```python
plt.figure(figsize=(10,8))
plt.subplot(121)
sns.barplot(x="month",y='DC',data=df_copy)

plt.subplot(122)
sns.barplot(x="day",y='DC',data=df_copy)
```

### DC is high for month(aug,sep,oct) and weekdays(tue,wed,thurs)
